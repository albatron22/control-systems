## 1. Импорт необходимых библиотек
Принято выносить импорты всех библиотек в первый блок кода файла Jupyter Notebook

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Генерация отсчетов времение с помощью функции `linspace` из библиотеки `numpy`.

```python
T = np.linspace(0.1, 10, 300)
```

# 3. Вычисление значений функции
$$y(t)=5 \ln(t^2) + 10$$

```python
y = 5.0 * np.log(T**2) + 10.0
```

Можно вывести значения массива `y` через `print`:

```python
print(y)
```

## 4. Строим график функции
$$y(t)=5 \ln(t^2) + 10$$

```python
plt.plot(T, y)
plt.grid(True) # включение сетки
plt.title('$y(t) = 5 \ln(t^2) + 10$')
plt.xlabel('time') # подпись оси абсцисс - оси времени
plt.ylabel('y(t)') # подпись оси ординат - ось значений функции
```
